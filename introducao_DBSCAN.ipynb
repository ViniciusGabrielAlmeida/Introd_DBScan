```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

# Criando dados


```python
from sklearn.datasets import make_moons

X, y_true = make_moons(n_samples=1000,noise=0.07, random_state=133)
"""X → Contém as características (features) dos dados.
É uma matriz NumPy com 1000 linhas e 2 colunas (representando coordenadas x e y dos pontos no plano 2D).

y_true → Contém os rótulos (classes) de cada ponto.
É um vetor com 1000 valores, sendo 0 ou 1 indicando a qual "lua" o ponto pertence."""
# noise adiciona ruído gaussiano aos dados, tornando os limites menos exatos.

plt.scatter(X[:,0],X[:,1], s=50);
```

# Aplicando Kmeans para comparação

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:,0], X[:,1],s=10, c=y_kmeans, cmap="rainbow")
#(dados eixo X, dados eixo y, tamanho, divisão, cores)
```

# Aplicando DBScan

##### O DBSCAN (Density-Based Spatial Clustering of Applications with Noise) é um algoritmo de clusterização não supervisionado que agrupa pontos com base na densidade dos dados. Ele identifica regiões densas como clusters, conectando pontos que estão próximos entre si (dentro de uma distância chamada epsilon) e que possuem um número mínimo de vizinhos (minPts). Diferente do k-means, o DBSCAN não exige definir o número de clusters previamente e consegue detectar grupos de formas variadas, além de identificar pontos isolados como ruído ou outliers. É especialmente eficaz para dados com formas complexas e presença de ruídos.

```python
from sklearn.cluster import DBSCAN

DBscan = DBSCAN(eps=0.08, min_samples=2)
#colocando parametros -> eps = Distância entre os centroides, min_samples = minimo de vizinhos para ser socre point
DBscan.fit(X)
#Inserindo dados
y_dbscan = DBscan.fit_predict(X)
#criando clusters

plt.scatter(X[:,0], X[:,1],s=10, c=y_dbscan, cmap="rainbow")
#(dados eixo X, dados eixo y, tamanho, cores)
```
